```python
import pandas as pd
import numpy as np
import nltk
from nltk import pos_tag
from nltk.corpus import wordnet as wn
```

```python
emotion = pd.read_csv('sample_words (manual exclusions).csv')
```

```python
# Ensure you have the required NLTK data files
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')
```

```python
# Define a function to detect past tense and plural forms
def is_past_tense_or_plural(word, pos_tag):
    if pos_tag.startswith('VBD') or pos_tag.startswith('VBN'):  # Past tense verbs
        return True
    if pos_tag.startswith('NNS'):  # Plural nouns
        return True
    return False

# Tag the words with POS tags
emotion['pos_tag'] = emotion['word'].apply(lambda x: pos_tag([x])[0][1])

# Filter the DataFrame
df_filtered = emotion[~emotion.apply(lambda row: is_past_tense_or_plural(row['word'], row['pos_tag']), axis=1)]
```
